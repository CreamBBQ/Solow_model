```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
eu = pd.read_csv('eu.csv')
col = pd.read_csv('col.csv')
```

```python
Y = col[col.VariableCode == 'cgdpo'][['YearCode','AggValue']].rename(columns={'AggValue': 'Y'}).set_index('YearCode')
N = col[col.VariableCode == 'pop'][['YearCode','AggValue']].rename(columns={'AggValue': 'N'}).set_index('YearCode')
K = col[col.VariableCode == 'cn'][['YearCode','AggValue']].rename(columns={'AggValue': 'K'}).set_index('YearCode')
L = col[col.VariableCode == 'emp'][['YearCode','AggValue']].rename(columns={'AggValue': 'L'}).set_index('YearCode')
prod = col[col.VariableCode == 'ctfp'][['YearCode','AggValue']].rename(columns={'AggValue': 'prod'}).set_index('YearCode')
df_col = pd.concat([Y,N,K,L,prod], axis = 1)
df_col['year'] = range(1950, 2020)
df_col.to_csv(r'clear_col.csv', index=False)
```

```python
Y = eu[eu.VariableCode == 'cgdpo'][['YearCode','AggValue']].rename(columns={'AggValue': 'Y'}).set_index('YearCode')
N = eu[eu.VariableCode == 'pop'][['YearCode','AggValue']].rename(columns={'AggValue': 'N'}).set_index('YearCode')
K = eu[eu.VariableCode == 'cn'][['YearCode','AggValue']].rename(columns={'AggValue': 'K'}).set_index('YearCode')
L = eu[eu.VariableCode == 'emp'][['YearCode','AggValue']].rename(columns={'AggValue': 'L'}).set_index('YearCode')
df_eu = pd.concat([Y,N,K,L], axis = 1)
df_eu['year'] = range(1950,2020)
df_eu.to_csv(r'clear_eu.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.lineplot(df_col.index.values, df_col['Y'])
ax.set_title('Salinity over Time', fontsize = 20, loc='center', fontdict=dict(weight='bold'))
ax.set_xlabel('Year', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('Salinity', fontsize = 16, fontdict=dict(weight='bold'))
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
```
