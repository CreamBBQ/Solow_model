```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.6
s = 0.8
delta = 0.4
n = 0.1
k_ast = (s/(n+delta))**(1/(1-alpha))
y_ast = (s/(n+delta))**(alpha/(1-alpha))
```

```python
k = np.linspace(0,100,1000)
y = k**alpha
save = s*y
dep = (n+delta)*k
fig = plt.figure(figsize=(7,7))
plt.plot(k,y, label = 'y', color = 'blue')
plt.plot(k,save, label = 'sy', color = 'orange')
plt.plot(k,dep, label = '(n+Î´)k', color = 'green')
plt.axvline(x=k_ast, color='k', linestyle='--', label = 'k*')
plt.axhline(y=y_ast, color='k', linestyle='-.', label ='y*')
plt.xlabel('k')
plt.ylim(0, 2.5)
plt.xlim(0,4)
plt.legend()
plt.xlabel('k')
plt.ylabel('y')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Aumento de s 

```python
s_prima = 0.9
k_ast_prima = (s_prima/(n+delta))**(1/(1-alpha))
y_ast_prima = (s_prima/(n+delta))**(alpha/(1-alpha))
k = np.linspace(0,100,1000)
y = k**alpha
save = s*y
save_prima = s_prima*y
dep = (n+delta)*k
fig = plt.figure(figsize=(7,7))
plt.plot(k,save, label = 'sy', color = 'gray')
plt.plot(k,save_prima, label = "s'y", color = 'orange')
plt.plot(k,dep, label = '(n+Î´)k', color = 'green')
plt.axvline(x=k_ast, color='gray', linestyle='--', label = 'k*')
plt.axhline(y=y_ast, color='gray', linestyle='-.', label ='y*')
plt.axvline(x=k_ast_prima, color='k', linestyle='--', label = "k*'")
plt.axhline(y=y_ast_prima, color='k', linestyle='-.', label ="y*'")
plt.xlabel('k')
plt.ylim(0, 2.5)
plt.xlim(0,5)
plt.legend()
plt.xlabel('k')
plt.ylabel('y')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Aumento de n

```python
n_prima = 0.3
k_ast_prima = (s/(n_prima+delta))**(1/(1-alpha))
y_ast_prima = (s/(n_prima+delta))**(alpha/(1-alpha))
k = np.linspace(0,100,1000)
y = k**alpha
save = s*y
dep = (n+delta)*k
dep_prima = (n_prima+delta)*k
fig = plt.figure(figsize=(7,7))
plt.plot(k,save, label = 'sy', color = 'orange')
plt.plot(k,dep, label = '(n+Î´)k', color = 'gray')
plt.plot(k,dep_prima, label = '(n+Î´)k', color = 'green')
plt.axvline(x=k_ast, color='gray', linestyle='--', label = 'k*')
plt.axhline(y=y_ast, color='gray', linestyle='-.', label ='y*')
plt.axvline(x=k_ast_prima, color='k', linestyle='--', label = "k*'")
plt.axhline(y=y_ast_prima, color='k', linestyle='-.', label ="y*'")
plt.xlabel('k')
plt.ylim(0, 2.5)
plt.xlim(0,5)
plt.legend()
plt.xlabel('k')
plt.ylabel('y')
plt.xticks([])
plt.yticks([])
plt.show()
```
