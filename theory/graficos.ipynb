```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.6
s = 0.8
delta = 0.4
n = 0.1
k_ast = (s/(n+delta))**(1/(1-alpha))
y_ast = (s/(n+delta))**(alpha/(1-alpha))
```

```python
k = np.linspace(0,100,1000)
y = k**alpha
save = s*y
dep = (n+delta)*k
fig = plt.figure(figsize=(7,7))
plt.plot(k,y, label = 'y', color = 'blue')
plt.plot(k,save, label = 'sy', color = 'orange')
plt.plot(k,dep, label = '(n+δ)k', color = 'green')
plt.axvline(x=k_ast, color='k', linestyle='--', label = 'k*')
plt.axhline(y=y_ast, color='k', linestyle='-.', label ='y*')
plt.xlabel('k')
plt.ylim(0, 2.5)
plt.xlim(0,4)
plt.legend()
plt.xlabel('k')
plt.ylabel('y')
plt.xticks([])
plt.yticks([])
#plt.gca().axes.get_yaxis().set_visible(False)
#plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
